## 1. Import useful libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
```

## 2. Read csv file

```python
df = pd.read_csv('Robbery_Open_Data.csv')
```

```python
df.shape
```

## 3. Check for null values and clean them

```python
df.isnull().sum()
```

Select null values with a mask and grab the entries with null values

```python
null_mask = df.isnull().any(axis=1)
entries_with_nulls = df[null_mask]
```

```python
entries_with_nulls[['OCC_YEAR', 'OCC_MONTH', 'OCC_DAY', 'OCC_DOY', 'OCC_DOW']]
```

Drop rows that have null values and reset the index

```python
df_clean = df.dropna(axis=0)
df_clean.reset_index(inplace=True)
df_clean.drop(columns='index', inplace=True)
```

Double-check if there are any remaining null values

```python
df_clean.isnull().sum()
```

## 4. Visualizing the data

Check the first 5 rows of the clean dataframe:

```python
df_clean.head()
```

Show the names of the columns we are working with:

```python
df_clean.columns
```

We use the unique function on the `OFFENCE` column to see how many robbery categories are in the dataset:

```python
df_clean['OFFENCE'].unique()
```

We want to see the number of reports for every type of robbery per year, therefore we group by those categories. We also select EVENT_UNIQUE_ID in order to apply the count function correctly.

```python
df_grouped = df_clean[['EVENT_UNIQUE_ID', 'REPORT_YEAR', 'OFFENCE']].groupby(['REPORT_YEAR', 'OFFENCE']).count()
```

Attempt 1: doesn't show the plot that we were expecting. The data isn't properly categorized.

```python
df_grouped.plot(kind='area', stacked=False, figsize=(8, 5), color='r')
plt.title('Wrong area plot')
```

Displaying some of the information inside `df_grouped`

```python
df_grouped.head(13)
```

If we look at df_grouped.index, we see that we have a multi index, we don't want that. We only want the year as the index, and the robbery time as columns.

```python
df_grouped.index
```

We can unstack the second level (index 1) of the multi index:

```python
df_area = df_grouped.unstack(level=1)
df_area.head()
```

Now, the index seems fine, however, lets see what happened to the columns:

```python
df_area.columns
```

We now have the multi index in the columns! As before, we want to fix this, therefore we eliminate the level of EVENT_UNIQUE_ID:

```python
# Drop the first level of the multi index
df_area.columns = df_area.columns.droplevel(level=0)

# Change the name of the columns category
df_area.columns.names = ['ROBBERY TYPE']

# Display first 5 rows
df_area.head()
```

Now, we plot the area chart:

```python
df_area.plot.area(figsize=(10, 10), stacked=True)
```

Looks fine, however there seems to be some weird things like why do all area plots for each category have the a similar shape? We decided to set the stacked argument to false to see how that changes the graph:

```python
df_area.plot.area(figsize=(10, 10), stacked=False)
```

Much better! Let's add some labels to the plot:

```python
ax = df_area.plot.area(figsize=(10, 10), stacked=False)
ax.set_title('Number of reports per category per year')
ax.set_xlabel('Year')
ax.set_ylabel('Reports')
plt.show()
```

### Easier way to do it:

Playing around with the code and searching on the internet for differente ways of doing the same things, we discovered this way of generating the same plots without having to deal with the multi-indices.

From the clean dataframe, group by `REPORT_YEAR` and `OFFENCE`, select the `OFFENCE` column and count the number of entries. Call `DataFrame.unstack()` to convert the each unique value in `OFFENCE` into a column.

```python
df_easy = df_clean.groupby(['REPORT_YEAR', 'OFFENCE'])['OFFENCE']\
            .count().\
            unstack(1)
```

As before, df_easy has the same information that we had in df_area. It has the number of robbery reports per year and for each unique robbery category found in the `OFFENCE` column. Below is the area plot showing the number of robbery reports per category by year:

```python
df_easy.plot.area(stacked=False, figsize=(10, 10))
ax.set_title('Number of reports per category per year')
ax.set_xlabel('Year')
ax.set_ylabel('Reports')
plt.show()
```

For the sake of visualizing the data in another way, we decided to make a bar plot with the same information:

```python
df_easy.plot.bar(figsize=(20, 20), stacked=True)
```

Lastly, we draw the information in df_easy as a line plot:

```python
df_easy.plot.line(figsize=(10, 10))
```

```python
comment_words = ''
stopwords = set(STOPWORDS)

# iterate through the csv file
for val in df_clean['OFFENCE']:

    # typecaste each val to string
    val = str(val)

    # split the value
    tokens = val.split()

    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()

    comment_words += " ".join(tokens)+" "

wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)

# plot the WordCloud image
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```
